```python
def get_color(r, g, b):
    """ Retourne une couleur à partir de ses composantes r, g, b entre 0 et 255"""
    return '#{:02x}{:02x}{:02x}'.format(r, g, b)

def draw_pixel(i,j,colors):
    tableau.create_rectangle((i,j)*2, outline= colors)

def ecran_aleatoire():
    couleurs = ["blue", "green", "black", "yellow", "magenta","red"]
    a = rd.choice(couleurs)
    for i in range(0,256):
        for j in range(0,256):
            a = rd.choice(couleurs)
            draw_pixel(i,j,a)

def degrade_gris():
    r = [i for i in range (0,256)]
    g = [i for i in range (0,256)]
    b = [i for i in range (0,256)]
    colors = []
    for i in range(0,256):
        colors.append(get_color(r[i],g[i],b[i]))
    for i in range (0,256):
        for j in range (0,256):
            draw_pixel(i,j,colors[i])

def degrade_2d():
    r = [i for i in range (0,256)]
    g = [i for i in range (0,256)]
    b = [i for i in range (0,256)]
    colors = []
    colors2 =[]
    for i in range(0,256):
        colors.append(get_color(r[i],0,b[-i]))
    for i in range (0,256):
        for j in range (0,256):
            draw_pixel(i,j,colors[i])

            
import random as rd
import tkinter as tk
Height = 256
Width = 256
dessin = tk.Tk()
dessin.title("degrade sah")
aleatoire = tk.Button(dessin,text = "aléatoire", command= ecran_aleatoire)
degra_gris = tk.Button(dessin,text="dégradé gris", command= degrade_gris)
degra_2D = tk.Button(dessin,text = "dégradé 2D", command= degrade_2d)
tableau = tk.Canvas(dessin, height = Height , width= Width ,bg= "black")
aleatoire.grid(column=0,row=0)
degra_gris.grid(column=0,row=1)
degra_2D.grid(column=0,row=2)
tableau.grid(column=1,row = 0,rowspan=3)
tableau.create_rectangle((10,10),(10,10), fill = "white")
dessin.mainloop()


```
