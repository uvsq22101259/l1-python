```python
def syracuse(n):
    """ Retourne la liste des valeurs de la suite en partant de n jusqu'à 1 """
    b = n
    n = []
    n.append(b)

    while b != 1:
       
        if b%2 == 0 :
            b = b / 2
        elif b%2 != 0 :
            b = b * 3 + 1
        n.append(b)

    return(n)


print(syracuse(6))
```

```python
def testeConjecture(n_max):
    """ Teste la conjecture de Collatz pour toutes les valeurs de 2 à n_max """
    for i in range(1 , n_max) :
        valable = syracuse(i)
        test = valable[-1] == 1
    return(test)



print(testeConjecture(10000))
```

```python
def syracuse2(n):
    """ Retourne la liste des valeurs de la suite en partant de n jusqu'à 1 et la valeurs du temps de vol """
    b = n
    n = []
    n.append(b)
    vol = 1
    b_max = 0

    while b != 1:
        vol += 1
        if b%2 == 0 :
            b = b / 2
        elif b%2 != 0 :
            b = b * 3 + 1
        n.append(b)
        if b > b_max :
            b_max = b
    return vol

def tempsVol(n):
    """ Retourne le temps de vol de n """
    n = syracuse2(n)
    return n

    
    

print("Le temps de vol de", 3, "est", tempsVol(3))
```

```python
def tempsVolListe(n_max):
    """ Retourne la liste de tous les temps de vol de 1 à n_max """
    for i in range(1 , n_max) :
         vol = syracuse2(i)
         print("temps de vol de ",i," est ",vol)
        
    return()


print(tempsVolListe(10000))
```

```python
def tv_max(n_max) :
    """retourne quel nombre a le plus grand temps de vol."""
    list_vol = []
    for i in range(1 , n_max) :
         vol = syracuse2(i)
         list_vol.append(vol)
        
    return(max(list_vol))
print(tv_max(10000))
    
```
